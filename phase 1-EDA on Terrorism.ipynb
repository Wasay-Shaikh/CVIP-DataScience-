## Import Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Load .CSV file


```python
df=pd.read_csv('globalterrorism.csv', encoding='latin-1')
df.head()
```

```python
df.shape
```

### Counting Columns with Missing Values in DataFrame 'df'



```python
nul = [i for i in df.columns if df[i].isna().sum() > 0]
len(nul)
```

### Remove Columns with Missing Values in DataFrame 'df'



```python
df.dropna(axis=1, inplace=True)
print(df.shape)
```

```python
df.info()
```

```python
df.describe()
```

## Number of Incidents each month

```python
sns.set_theme(style='darkgrid', font_scale=1.2)  # Increase the font size and set the style

plt.figure(figsize=(15, 7))
ax = sns.countplot(data=df, x='imonth')
plt.xticks(ha='right')  # Rotate x-axis labels and align them to the right
plt.xlabel('Month', labelpad=18)  # Add some padding to the x-axis label
plt.ylabel('Incident Count', labelpad=15)  # Add some padding to the y-axis label
plt.title('Incident Count Per Month (Overall)', pad=25)  # Add some padding to the title

# Annotate bars with counts
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.tight_layout()
plt.show()
```

### Number of Terror Activities Each Year


```python
plt.figure(figsize=(15, 8))
sns.set_theme(style='darkgrid', font_scale=1.2)  # Increase the font size

sns.countplot(x='iyear', data=df, edgecolor='black', palette='viridis')  # Use a different color palette
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels and align them to the right
plt.xlabel('Year of Attack', labelpad=15)  # Add some padding to the x-axis label
plt.ylabel('Number of Terrorist Activities', labelpad=15)  # Add some padding to the y-axis label
plt.title('Number of Terrorist Activities Each Year', pad=20)  # Add some padding to the title

plt.tight_layout()
plt.show()
```

## Terrorist Organisation with Highest Terror Attacks


```python
sns.set_theme(style='darkgrid', font_scale=1.2, palette='inferno')  # Increase the font size and set the style

plt.figure(figsize=(10, 8))
sns.barplot(x=df['gname'].value_counts()[1:16].index, y=df['gname'].value_counts()[1:16].values)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate x-axis labels and align them to the right
plt.xlabel('Terrorist Organization', labelpad=15)  # Add some padding to the x-axis label
plt.ylabel('Number of Terror Attacks', labelpad=10)  # Add some padding to the y-axis label
plt.title('Terrorist Organizations with Highest Terror Attacks', pad=20)  # Add some padding to the title

# Annotate bars with counts
for index, value in enumerate(df['gname'].value_counts()[1:16].values):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=8, color='black')

plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style='darkgrid', font_scale=1.2, palette='viridis')  # Increase the font size and set the style

plt.figure(figsize=(18, 12))
# Use the orient parameter to create a horizontal bar plot
ax = sns.barplot(x=df['region_txt'].value_counts().values, y=df['region_txt'].value_counts().index, orient='h')  
# Add some padding to the x-axis label
plt.xlabel('Number of Terror Attacks', labelpad=15)  
# Add some padding to the y-axis label
plt.ylabel('Region', labelpad=15)  
# Add some padding to the title
plt.title('Top Regions Affected by Terrorism', pad=20)

# Annotate bars with counts
for index, value in enumerate(df['region_txt'].value_counts().values):
    ax.text(value, index, str(value), ha='left', va='center', fontsize=13, color='black')

plt.tight_layout()
plt.show()
```
